```python
import pandas.util.testing as tm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import os

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df_ga = pd.read_csv('ga_traffic.csv')
lats = {}
lons = {}

cities = df_ga['City'].unique()
for city in cities:
    latitudes = np.array(df_ga[df_ga['City']==city]['Start_Lat']).reshape((-1, 1))
    longitudes = np.array(df_ga[df_ga['City']==city]['Start_Lng']).reshape((-1, 1))
    lats[city] = StandardScaler().fit(latitudes)
    lons[city] = StandardScaler().fit(longitudes)
df_ga['Start_Lat'] = df_ga.apply(lambda x: lats[x['City']]
                                .transform(np.array(x['Start_Lat']).reshape(-1,1))[0,0], axis=1)
df_ga['Start_Lng'] = df_ga.apply(lambda x: lons[x['City']]
                                .transform(np.array(x['Start_Lng']).reshape(-1,1))[0,0], axis=1)

df_ga.to_csv('./df_ga.csv',index=False)
```
